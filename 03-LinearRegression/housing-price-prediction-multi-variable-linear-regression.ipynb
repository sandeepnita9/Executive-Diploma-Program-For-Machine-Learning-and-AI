```python
#Now converting all categorical atrributes into binary 0 or 1
#we have following categorical attributes
categorical_attr = ['mainroad', 'basement', 'guestroom', 'hotwaterheating', 'airconditioning', 'prefarea']

df[categorical_attr] = df[categorical_attr].apply(lambda x : x.map({"yes":1, "no":0}))
df.head()
```

```python
#Lets deal with furnishingstatus column.
#Let's create dummy variable column.
furnishing_Status = pd.get_dummies(df.furnishingstatus)
furnishing_Status.head()
```

```python
#Logically we don't need 3 Columns and we can manage with two columns as well
furnishing_Status1 = pd.get_dummies(df.furnishingstatus, drop_first=True)
furnishing_Status1.head()
```

#00 means furnished</br>
#01 means unfurnished</br>
#10 means semi-furnished

```python
#lets join these dummy variable to main dataframe
df = pd.concat([df, furnishing_Status1], axis=1)
df.head()
```

```python
#furnishingstatus column is redundant now. Lets drop it
df = df.drop('furnishingstatus', axis=1)
df.head()
```

```python
#Split Data into Train and Test
df_train, df_test = train_test_split(df, train_size=0.7, random_state=100)
```

```python
#Scaling data using MinMax Approach

#first create object of min/Max Scaler class
scaler = MinMaxScaler()

#We will rescale only numerical columns and not the binary columns
num_var = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking', 'price']
```

```python
#lets fit and transform given variable
df_train[num_var] = scaler.fit_transform(df_train[num_var])
df_train.head()
```

## To Remove Columns from our Model ##
### We compare Significance level (p-value) and VIF (variance Inflation Factor (multicollinearity) ###

```python
#let's create a dataframe which will have all feature names and their VIF values
vif = pd.DataFrame()
vif['features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```

## Rules to follow to remove columns/features from Models

<h3> p-value greater than 0.05 considered HIGH</h3>
<h3> VIF greater than 5 considered HIGH </h3>

<h3> High p-value and High VIF  - Definitly Remove</h3>
<h3> High - Low Combination </h3>
<ol>
    <li>High p-value and Low VIF >>>  Remove these first</li>
    <li>Low p-value and High VIF >>>  Remove these Second</li>
</ol>
<h3> Low p-value and Low VIF </h3>

```python
#with above rule first we will remove features with High p-value
X = X_train.drop('semi-furnished' , axis=1)
```

```python
#Now lets create new Model without Furnished features
X_train_sm = sm.add_constant(X)

lr = sm.OLS(y_train, X_train_sm)
lr_model2 = lr.fit()
lr_model2.params
```

```python
lr_model2.summary()
```

```python
#let's create a dataframe which will have all feature names and their VIF values
vif = pd.DataFrame()
vif['features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```

```python
#based on the rule we checked above now lets remove bedroom as it has high p-value and high VIF both
#with above rule first we will remove features with High p-value
X = X.drop('bedrooms' , axis=1)
```

```python
#Now lets create new Model without Furnished features
X_train_sm = sm.add_constant(X)

lr = sm.OLS(y_train, X_train_sm)
lr_model2 = lr.fit()
lr_model2.params
```

```python
lr_model2.summary()
```

```python
#let's create a dataframe which will have all feature names and their VIF values
vif = pd.DataFrame()
vif['features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```
